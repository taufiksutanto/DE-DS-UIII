<center><h2><strong><font color="blue">Data Definition-Manipulation Languages</font></strong></h2></center>

## 1. Fundamental MariaDB/MySQL Data Types
Relational integrity depends on the selection of appropriate data types. Choosing the correct format optimizes storage and prevents data corruption:

### Numeric Types:
* **INT**: Standard integers for counts or IDs.
* **DECIMAL**: Necessary for financial or high-precision research data.
* **FLOAT/DOUBLE**: Used for approximate numerical values in scientific modeling.

### String Types:
* **VARCHAR(n)**: Variable-length strings (e.g., names or emails). Highly efficient.
* **TEXT**: Reserved for large blocks of descriptive research notes.
* **CHAR**: Fixed-length strings (e.g., ISO country codes).

### Temporal Types:
* DATE: Stores YYYY-MM-DD formats.
* DATETIME: Captures both date and time.
* TIMESTAMP: Automatically updates during record modification; ideal for audit trails.

<center><h2><strong><font color="blue">Data Definition Language (DDL) in MariaDB</font></strong></h2></center>

Data Definition Language (DDL) is the subset of SQL used to define, modify, and manage the **structure** of a database. Unlike Data Manipulation Language (DML), which handles the data inside the rows, DDL focuses on the "containers"—the databases, tables, and indexes themselves.

## 1. The CREATE Statement
The `CREATE` statement is the starting point for any database project. It is used to establish new databases, tables, indexes, or views.

### Creating a Database
```sql
CREATE DATABASE research_project;
```

### Creating a Table
When creating a table, you must define the column names, data types, and constraints (like `PRIMARY KEY`).
```sql
CREATE TABLE students (
    id INT AUTO_INCREMENT PRIMARY KEY,
    full_name VARCHAR(100) NOT NULL,
    email VARCHAR(255) UNIQUE,
    enrollment_date DATE DEFAULT CURRENT_DATE
);
```

> **Note:** Using `IF NOT EXISTS` is a best practice to prevent scripts from crashing if the table or database already exists.

## 2. The ALTER Statement
The `ALTER` statement is used to modify an existing structure without deleting it. This is useful when requirements change mid-project, such as needing to add a new column or change a data type.

### Common ALTER Operations:
* **Adding a Column:**
    ```sql
    ALTER TABLE students ADD COLUMN phone_number VARCHAR(15);
    ```
* **Modifying a Column Type:**
    ```sql
    ALTER TABLE students MODIFY COLUMN full_name VARCHAR(150);
    ```
* **Dropping a Column:**
    ```sql
    ALTER TABLE students DROP COLUMN phone_number;
    ```
* **Renaming a Table:**
    ```sql
    ALTER TABLE students RENAME TO registered_students;
    ```
    
## 3. The DROP Statement

The `DROP` statement is the "wrecking ball" of SQL. It is used to permanently delete a table or an entire database. 

> **Warning:** Use `DROP` with extreme caution. Once a table is dropped, all the data stored within it is lost forever unless you have a backup.

### Deleting a Table
```sql
DROP TABLE IF EXISTS registered_students;
```

### Deleting a Database
```sql
DROP DATABASE research_project;
```

<center><h2><strong><font color="blue">Data Manipulation Language (DML) in MariaDB</font></strong></h2></center>
While Data Definition Language (DDL) builds the "house," **Data Manipulation Language (DML)** is the act of moving furniture in. It allows researchers to interact with the actual data stored within the tables. In the context of the **CRUD** (Create, Read, Update, Delete) cycle, DML is where the majority of your daily data engineering work occurs.

## 1. The SELECT Statement (Reading Data)
The `SELECT` statement is the most frequent operation in data science. It retrieves specific records from one or more tables based on defined criteria.

### Basic Syntax
```sql
SELECT full_name, email 
FROM students 
WHERE country_origin = 'Indonesia' 
ORDER BY full_name ASC;
```

## 2. The INSERT Statement (Creating Data)
`INSERT` adds new rows to a table. For production-grade research, we often handle this through Python scripts to ensure data is sanitized.

### Standard Insertion
```sql
INSERT INTO instructors (full_name, email) 
VALUES ('Dr. Aris Kusuma', 'aris.k@uiii.ac.id');
```

### Multi-row Insertion
```sql
INSERT INTO students (full_name, country_origin) 
VALUES 
    ('Siti Aminah', 'Indonesia'),
    ('John Doe', 'Australia'),
    ('Li Wei', 'China');
```

## 3. The UPDATE Statement (Modifying Data)
The `UPDATE` statement modifies existing records. In a professional environment, this must be handled with extreme precision.

### Critical Usage
```sql
UPDATE enrollments 
SET grade = 3.85 
WHERE student_id = 101 AND course_id = 5;
```

## 4. The DELETE Statement (Removing Data)
The `DELETE` statement removes specific rows from a table. Unlike the `DROP` command (which removes the table itself), `DELETE` only removes the data.

### Targeted Deletion
```sql
DELETE FROM students 
WHERE enrollment_date < '2020-01-01';
```

## DML Command Summary
| Command | CRUD Equivalent | Research Utility |
| :--- | :--- | :--- |
| **SELECT** | Read | Extracting features for ML models or EDA. |
| **INSERT** | Create | Ingesting sensor data or web-scraped results. |
| **UPDATE** | Update | Cleaning "dirty" data or updating model logs. |
| **DELETE** | Delete | Removing outliers or outdated experimental runs. |

<center><h2><strong><font color="blue">WHERE Clause Variants in MySQL Data Manipulation Language (DML)</font></strong></h2></center>

In MySQL Data Manipulation Language (DML), the `WHERE` clause serves as a fundamental mechanism for restricting the scope of operations such as `SELECT`, `UPDATE`, and `DELETE`. It enables conditional filtering of tuples (rows) based on logical predicates, thereby ensuring that only relevant data is retrieved or modified.

Formally, the `WHERE` clause evaluates a Boolean expression for each row in a relation, returning only those rows for which the condition evaluates to `TRUE`.

### 2. Basic Syntax
```sql
SELECT column_list
FROM table_name
WHERE condition;
```

The `condition` is a logical expression composed of operators, constants, column references, and functions.

### 3. Variants of WHERE Clause Conditions
The expressive power of the `WHERE` clause arises from its support for multiple types of conditions. These variants are outlined below.

### 3.1 Comparison Operators
Comparison operators are used to compare column values with constants or other columns.

| Operator | Description              |
| -------- | ------------------------ |
| `=`      | Equal to                 |
| `<>`     | Not equal to             |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |

**Example:**
```sql
SELECT * 
FROM employees
WHERE salary > 5000;
```

### 3.2 Logical Operators
Logical operators combine multiple conditions.

| Operator | Description         |
| -------- | ------------------- |
| `AND`    | Logical conjunction |
| `OR`     | Logical disjunction |
| `NOT`    | Logical negation    |

**Example:**
```sql
SELECT * 
FROM employees
WHERE department = 'IT' AND salary > 5000;
```

### 3.3 BETWEEN Operator
The `BETWEEN` operator specifies a range of values (inclusive).
**Example:**
```sql
price >= 100 AND price <= 500
```

### 3.4 IN Operator
The `IN` operator allows matching against a set of discrete values.
**Example:**
```sql
SELECT * 
FROM students
WHERE major IN ('Mathematics', 'Physics', 'Statistics');
```

### 3.5 LIKE Operator (Pattern Matching)

The `LIKE` operator is used for string pattern matching.

| Symbol | Meaning                 |
| ------ | ----------------------- |
| `%`    | Zero or more characters |
| `_`    | Exactly one character   |

**Example:**
```sql
SELECT * 
FROM customers
WHERE name LIKE 'A%';
```

### 3.6 IS NULL Operator
This operator checks for `NULL` values, which represent missing or undefined data.
**Example:**

```sql
SELECT * 
FROM orders
WHERE shipped_date IS NULL;
```

### 3.7 EXISTS Operator
The `EXISTS` operator tests for the existence of rows returned by a subquery.
**Example:**
```sql
SELECT * 
FROM departments d
WHERE EXISTS (
    SELECT 1 
    FROM employees e 
    WHERE e.department_id = d.department_id
);
```

### 3.8 Subqueries in WHERE Clause
The `WHERE` clause can incorporate subqueries for dynamic filtering.
**Example:**
```sql
SELECT * 
FROM employees
WHERE salary > (
    SELECT AVG(salary) 
    FROM employees
);
```

<center><h1><strong><font color="blue">Advanced Selection: Joins (Inner, Left, Right, Full)</font></strong></h1></center>
In relational database systems such as MySQL, a **JOIN** operation is used to combine tuples (rows) from two or more relations (tables) based on a related attribute. JOINs operationalize the relational algebra concept of combining relations through common keys, thereby enabling normalized database schemas to be queried in a logically unified manner.

### 2. General Syntax
```sql
SELECT column_list
FROM table1
JOIN table2
ON table1.common_column = table2.common_column;
```
The `ON` clause specifies the join predicate, typically based on primary key–foreign key relationships.

### 3. Types of JOINs
* **INNER JOIN**: Returns only rows with matching values in both tables.
* **LEFT JOIN**: Returns all rows from the left table and matched rows from the right table. Non-matching rows are filled with `NULL`.
* **RIGHT JOIN**: Returns all rows from the right table and matched rows from the left table.
* **OUTER JOIN**: Returns all rows from both tables, with `NULL` for non-matching records.
* **SELF JOIN**: A table is joined with itself, often used for hierarchical or recursive relationships.
```sql
SELECT a.name AS employee, b.name AS manager
FROM employees a
JOIN employees b
ON a.manager_id = b.employee_id;
```

<center><h1><strong><font color="blue">Aggregations in MySQL Databases</font></strong></h1></center>
In MySQL, **aggregation** refers to the process of computing summary statistics over a set of tuples (rows). Aggregation operations are fundamental in analytical querying, enabling the transformation of raw data into meaningful summaries such as totals, averages, counts, and extrema. These operations are typically realized *aggregate functions* applied to one or more columns.

### 2. Common Aggregate Functions
MySQL provides several built-in aggregate functions:

| Function  | Description                                  |
| --------- | -------------------------------------------- |
| `COUNT()` | The number of rows or non-`NULL` values |
| `SUM()`   | Computes the total sum of numeric values     |
| `AVG()`   | Computes the average value                   |
| `MIN()`   | Returns the minimum value                    |
| `MAX()`   | Returns the maximum value                    |

**Example:**

```sql
SELECT COUNT(*) AS total_employees,
       AVG(salary) AS avg_salary
FROM employees;
```

### 3. GROUP BY Clause
Aggregation is often used in conjunction with the `GROUP BY` clause, which partitions the relation into disjoint groups based on specified attributes. Aggregate functions are then applied independently to each group.

```sql 
SELECT department_id, AVG(salary) AS avg_salary
FROM employees
GROUP BY department_id;
```

### 4. HAVING Clause
The `HAVING` clause filters groups based on aggregate conditions, analogous to how `WHERE` filters individual rows.

```sql
SELECT department_id, COUNT(*) AS num_employees
FROM employees
GROUP BY department_id
HAVING COUNT(*) > 5;
```

### 5. DISTINCT in Aggregation

Aggregate functions can incorporate the `DISTINCT` keyword to eliminate duplicate values prior to computation.

```sql
SELECT COUNT(DISTINCT department_id) AS unique_departments
FROM employees;
```

### 6. NULL Handling
Aggregate functions generally ignore `NULL` values, except for `COUNT(*)`, which counts all rows regardless of `NULL` presence.

<center><h1><strong><font color="blue">Normalizations in MySQL Databases</font></strong></h1></center>

**Normalization** is a systematic design methodology in relational database theory aimed at reducing redundancy and eliminating undesirable characteristics such as update, insertion, and deletion anomalies. It involves decomposing relations (tables) into smaller, well-structured relations while preserving data integrity and functional dependencies.

The normalization process is guided by a sequence of *normal forms*. This discussion is limited to the first three: **First Normal Form (1NF)**, **Second Normal Form (2NF)**, and **Third Normal Form (3NF)**.

### 2. First Normal Form (1NF)
A relation is in **First Normal Form (1NF)** if:
1. All attributes contain **atomic (indivisible) values**.
2. There are **no repeating groups or multivalued attributes**.
3. Each record can be uniquely identified (i.e., existence of a primary key).

**Example (Not in 1NF):**
```sql
StudentID | Name   | Courses
----------|--------|------------------------
101       | Alice  | Math, Physics
```
**Transformation to 1NF:**

```sql id="y6d1kp"
StudentID | Name   | Course
----------|--------|---------
101       | Alice  | Math
101       | Alice  | Physics
```

### 3. Second Normal Form (2NF)
A relation is in **Second Normal Form (2NF)** if:
1. It is already in **1NF**.
2. All non-key attributes are **fully functionally dependent** on the entire primary key (i.e., no **partial dependency** on a subset of a composite key).

**Example (Not in 2NF):**
```sql
(StudentID, CourseID) → Instructor
CourseID → CourseName
```
Here, `CourseName` depends only on `CourseID`, not the full composite key.

**Decomposition:**
```sql
Courses(CourseID, CourseName)
Enrollments(StudentID, CourseID, Instructor)
```

### 4. Third Normal Form (3NF)
A relation is in **Third Normal Form (3NF)** if:
1. It is already in **2NF**.
2. There are **no transitive dependencies**, i.e., non-key attributes do not depend on other non-key attributes.

**Example (Not in 3NF):**
```sql 
EmployeeID → DepartmentID
DepartmentID → DepartmentName
```
Here, `DepartmentName` is transitively dependent on `EmployeeID`.

**Decomposition:**

```sql id="b3n6xp"
Employees(EmployeeID, DepartmentID)
Departments(DepartmentID, DepartmentName)
```

## Legal Aspects of Digital Data Gathering for Research
### 1. Introduction
Digital data gathering for research refers to the collection of data from digital sources such as online platforms, databases, sensors, and information systems. From a legal perspective, such activities are governed by a combination of **data protection laws**, **privacy regulations**, **intellectual property rights**, and **terms of service agreements**. Compliance with these legal frameworks is essential to ensure that data collection practices are lawful, accountable, and defensible.

This discussion focuses strictly on **legal compliance** (not ethical considerations), with emphasis on both **Indonesian regulations** and **general international frameworks**.


### 2. Legal Foundations in Digital Data Collection
#### 2.1 Lawful Basis for Data Processing
A central legal requirement in most jurisdictions is the existence of a **lawful basis** for collecting and processing data. Common legal bases include:

* Explicit consent from the data subject
* Contractual necessity
* Legal obligation
* Legitimate interest (subject to regulatory constraints)

Researchers must clearly identify and document the legal basis before initiating data collection.


#### 2.2 Data Protection and Privacy Laws
##### (a) Indonesian Context
In Indonesia, digital data collection is primarily regulated by:
* **Law No. 27 of 2022 on Personal Data Protection (PDP Law)**
  * Establishes principles for lawful processing of personal data
  * Requires consent for data collection
  * Defines rights of data subjects (e.g., access, correction, deletion)

* **Law No. 11 of 2008 on Electronic Information and Transactions (ITE Law)** (as amended)
  * Regulates electronic systems and digital interactions
  * Includes provisions on unauthorized access and data misuse

* Implementing regulations governing electronic system providers (PSE)

Researchers must ensure that any collection of **personally identifiable information (PII)** complies with these statutes.


##### (b) International (General) Context
Globally, several influential regulatory frameworks shape digital data practices:
* **General Data Protection Regulation (GDPR) – European Union**
  * Strong emphasis on consent, transparency, and accountability
  * Introduces principles such as data minimization and purpose limitation
  * Applies extraterritorially in certain cases

* **California Consumer Privacy Act (CCPA)** (and CPRA) – United States
  * Provides rights to access, delete, and opt out of data selling

* **OECD Privacy Guidelines**
  * Establish internationally recognized principles for data protection

Even when research is conducted outside these jurisdictions, compliance may still be required if the data involves individuals from those regions.

### 3. Data Classification and Sensitivity
Legal obligations vary depending on the **type of data collected**:
* **Personal Data:** Identifiable individual information (e.g., name, email)
* **Sensitive Personal Data:** Higher-risk categories (e.g., health, biometrics)
* **Non-Personal Data:** Aggregated or anonymized data

Stricter legal requirements apply to sensitive data, including enhanced protection measures and, in some cases, explicit consent.


### 4. Data Collection Methods and Legal Constraints
Different methods of digital data gathering introduce distinct legal considerations:
* **Web Scraping:** Must comply with website terms of service and may implicate intellectual property and anti-hacking laws
* **APIs (Application Programming Interfaces):** Governed by licensing agreements and usage policies
* **Surveys and Online Forms:** Require clear consent mechanisms and privacy notices
* **Sensor and Log Data Collection:** Must ensure no unauthorized surveillance or data interception

Failure to comply may result in legal liability, including civil penalties or criminal sanctions.

<center><h1><strong><font color="blue">Ethics in Digital Data Gathering for Research</font></strong></h1></center>

While legal compliance establishes the **minimum mandatory standards** for digital data gathering, **ethical considerations** extend beyond legality to address issues of fairness, respect, and responsibility toward data subjects and society. In research contexts, particularly within data science, ethical practice ensures that data-driven inquiry does not harm individuals, communities, or institutions, even when such actions may be legally permissible.

This section builds upon prior legal discussion by positioning ethics as a **complementary and often stricter framework** guiding responsible data use in both Indonesian and international contexts.


### 2. Relationship Between Legal Compliance and Ethics

Legal and ethical frameworks are related but not identical:
* **Law** defines what is *permissible*
* **Ethics** defines what is *appropriate and responsible*

A legally compliant activity may still be ethically problematic. For example, collecting publicly available personal data may comply with regulations but still violate expectations of privacy or contextual integrity.

**Conceptual Distinction:**

| Dimension   | Legal Perspective             | Ethical Perspective                         |
| ----------- | ----------------------------- | ------------------------------------------- |
| Basis       | Statutory/regulatory rules    | Moral principles and professional standards |
| Enforcement | External (courts, regulators) | Internal (research integrity, peer review)  |
| Scope       | Minimum requirement           | Often broader and stricter                  |


### 3. Core Ethical Principles in Digital Data Research
#### 3.1 Respect for Persons (Autonomy)

Researchers must respect the autonomy and dignity of individuals whose data is being used.
**Practical Implications:**
* Ensure meaningful informed consent (not merely formal compliance)
* Avoid deceptive data collection practices

#### 3.2 Beneficence and Non-Maleficence

Derived from biomedical research ethics, these principles require:
* Maximizing potential benefits
* Minimizing potential harm

**Examples of Harm:**
* Re-identification of anonymized data
* Misuse of sensitive attributes (e.g., religion, health status)
* Algorithmic bias leading to unfair outcomes

#### 3.3 Justice and Fairness
Data-driven research must ensure equitable treatment across populations.

**Concerns:**
* Sampling bias
* Underrepresentation of vulnerable groups
* Discriminatory model outcomes

#### 3.4 Transparency and Accountability

Researchers must be transparent about:

* Data sources
* Data processing methods
* Limitations and risks

Accountability requires that decisions can be audited and justified.


### 4. Ethical Issues in Common Data Science Practices
#### 4.1 Social Media Data Mining

* Ethical issue: Users may not expect their data to be used for research
* Even if legally public, contextual privacy expectations must be considered

#### 4.2 Web Scraping
* Ethical issue: Large-scale extraction may burden systems or violate implicit norms
* Respect for platform intent and user expectations is critical


#### 4.3 Machine Learning on Personal Data
* Ethical issue: Bias, profiling, and unintended inference
* Requires fairness-aware modeling and validation


### 5. Indonesian Context
In Indonesia, ethical governance in research is typically institutionalized through:

* **Research Ethics Committees (Komite Etik Penelitian)** at universities
* Guidelines issued by:
  * Ministry of Education, Culture, Research, and Technology
  * Institutional review boards (IRB-equivalent bodies)

Although laws such as the **Personal Data Protection Law (UU No. 27/2022)** provide legal structure, ethical oversight ensures:
* Responsible interpretation of consent
* Protection of vulnerable populations
* Context-sensitive data usage

For institutions such as Universitas Islam Internasional Indonesia, postgraduate research is generally expected to undergo ethical clearance when involving human-related data.

### 6. International (General) Context

Globally, ethical standards are shaped by widely recognized frameworks:
* **Belmont Report**
  * Respect for persons
  * Beneficence
  * Justice

* **Association for Computing Machinery Code of Ethics**
  * Emphasizes responsibility in computing and data use

* Institutional Review Boards (IRBs) / Ethics Committees
  * Mandatory in many jurisdictions for human-subject research

These frameworks often extend beyond legal requirements, particularly in emerging areas such as artificial intelligence and big data analytics.

### 7. Practical Ethical Implementation in Data Science Research
Students and researchers should operationalize ethics through:

#### 7.1 Ethical Impact Assessment
* Identify potential risks to individuals and groups
* Evaluate downstream consequences of data use

#### 7.2 Data Minimization with Ethical Justification
* Not only “what is legal to collect” but “what is necessary and justified”

#### 7.3 Anonymization and Risk Awareness
* Recognize limits of anonymization (risk of re-identification)

#### 7.4 Documentation

Include an **Ethical Considerations Section**:

```markdown
## Ethical Considerations
This study uses publicly available data; however, all user identifiers have been anonymized to prevent re-identification. Potential biases in the dataset are acknowledged and mitigated through balanced sampling.
```

### 8. Common Ethical Pitfalls

* Treating public data as ethically unrestricted
* Ignoring potential harm from model outputs
* Overlooking bias and representational issues
* Reducing ethics to a “checkbox” rather than an ongoing process
